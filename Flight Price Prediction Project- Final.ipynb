### EDA 

```python
# Splitting Departure time into hour and minute:

df["Departure_time"] = pd.to_datetime(df["Departure_time"])

df["dep_hour"] = df["Departure_time"].dt.hour

df["dep_min"] = df["Departure_time"].dt.minute

df.drop(columns = ['Departure_time'],axis =1, inplace = True)        # Dropping the Departure_time column after splitting
```

```python
# Splitting Arrival time into hour and minute:

df["Arrival_time"] = pd.to_datetime(df["Arrival_time"])
df['Arr_hour'] = df["Arrival_time"].dt.hour
df['Arr_min'] = df["Arrival_time"].dt.minute

df.drop(columns = ['Arrival_time'],axis =1, inplace = True)          # Dropping the Departure_time column after splitting
```

```python
# Splitting date into day and month as year column is same for all:

df["Date"]= pd.to_datetime(df["Date"])
df["Dt_Day"]=df["Date"].dt.day
df["Dt_Month"]=df["Date"].dt.month

df.drop(columns = ['Date'],axis =1, inplace = True)                   # Dropping the Departure_time column after splitting
```

```python
# Splitting duration column into hours and minutes in seperate column :

t = pd.to_timedelta(df['Duration'])

df['Dur_hrs']=t.dt.components['hours']
df['Dur_mins']=t.dt.components['minutes']

df.drop(["Duration"], axis=1, inplace = True)
```

```python
# Changing the non in the No of stop column into 0 and changing its type to int

df["No of stop"] = df["No of stop"].replace("Non",0)

df["No of stop"]= df["No of stop"].astype("int64")
```

### Checking the skewness :

```python
plotnumber=1
plt.figure(figsize=(15,40))
for i in df:
    if plotnumber < 22:
        ax=plt.subplot(11,2,plotnumber)
        sns.kdeplot(df[i],data=df,color="g")
        plt.xlabel(i,fontsize=15)
        plotnumber+=1
plt.show()
```

```python
df.skew()
```

#### We some skewness is present in Dur_hrs column rest every continuous column is within the range of +/- 0.5, price is our target variable hence we do not have to remove the skewness from it and the rest are all categorical column hence we will not consider their skewness.

```python
# Using Power Transform to remove skewness in dur_hrs column:

from sklearn.preprocessing import PowerTransformer

transform = PowerTransformer()
df[["Dur_hrs", "Dt_day"]]= transform.fit_transform(df[["Dur_hrs", "Dt_Day"]])
```

```python
df.skew()
```

#### After appliying the power transformer the skewness has decreased to the large extent and is now between the desired value.

## Model Building :

```python
#splitting the dataset: 

x=df.drop(columns=["Price",],axis=1)
y=df["Price"]
```

```python
# Importing the necessary libraries for model building :

from sklearn.linear_model import LinearRegression,Lasso
from xgboost import XGBRegressor
from sklearn.ensemble import GradientBoostingRegressor,BaggingRegressor,RandomForestRegressor,AdaBoostRegressor
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split,cross_val_score,GridSearchCV,KFold
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
```

```python
# scaling the dataset using standard scaler :

scale=StandardScaler()
x_scaled=scale.fit_transform(x)
```

### Linear Regression Model Building :

#### Finding the best random state of linear regression :

```python
lr = LinearRegression()

for i in range(0,100):
    x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size =0.20, random_state =i)
    lr.fit(x_train, y_train)
    pred_train = lr.predict(x_train)
    pred_test = lr.predict(x_test)
    print(f"At random state {i}, the training accuracy is : {r2_score(y_train, pred_train)}" )
    print(f"At random state {i}, the testing accuracy is : {r2_score(y_test, pred_test)}" )
    print('\n')
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.20,random_state=84)
```

```python
k=KFold(n_splits=5,random_state=84,shuffle=True)
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
x =r2_score(y_test,y_pred)
```

```python
cross = cross_val_score(lr,x_scaled,y,cv=k).mean()
```

```python
print("r-2 score : ",x,"\n","cross validation score :",cross)
```
